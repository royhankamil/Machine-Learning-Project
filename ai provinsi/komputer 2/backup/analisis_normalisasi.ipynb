```python
from wordcloud.wordcloud import WordCloud


def showing(wc):
    plt.figure(figsize=(24,60))
    plt.axis("off")
    plt.imshow(wc, interpolation="bilinear")
    plt.show

allword = "".join([text for text in df.comment.to_list() if type(text) != float])
wc = WordCloud()
img = wc.generate(allword)

word_cloud = WordCloud(
    height=1200,
    width=800,
    collocations=False,
    colormap="Blues_r"
 )

showing(img)
```
